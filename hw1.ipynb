```python
import numpy as np 


def f(x):
    return x**4 / (x**6 + 1)

N = [np.floor(10**(1+0.1*i)) for i in range(41)]
points = []
estimates = []
for n in N:
    count = 0
    for i in range(int(n)):
        point = np.random.random(2)
        if point[1] < f(point[0]):
            count += 1
    estimates.append(count/n)

# Analytical answer
ans = (
    -np.log(np.sqrt(3) + 2) / (4 * np.sqrt(3))
    + np.arctan(np.sqrt(3) + 2) / 6
    - np.arctan(np.sqrt(3) - 2) / 6
    + np.log(2 - np.sqrt(3)) / (4 * np.sqrt(3))
    + np.pi / 12
)

import matplotlib.pyplot as plt

plt.axhline(ans, color="r", label="Analytical Solution")
plt.xscale("log")
plt.plot(N, estimates, label="Monte Carlo")
plt.title(r"Value of $\int_0^1 \frac{u^4}{u^6+1}du$")
plt.legend()
```

```python
def pdf(x):
    return 3*x**2

T = []
for i in range(10**5):
    arrival = np.random.random(3)
    t = max(arrival)
    T.append(t)

x = np.linspace(0, 1, 1000)

plt.plot(x, pdf(x), label="Theoretical PDF")
plt.hist(T, bins=100, density=T, label="Simulated arrival time")
plt.legend()
```
