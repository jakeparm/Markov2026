1.d)

```python
import numpy as np

def f(x):
    return x*(1+x)*np.e**(-x)/3

def c(a):
    return 1/(3*a**2)*1/(1-a)*np.e**(-a)

def g(a, x):
    return a**2*x*np.e**(-a*x)

a_star = np.sqrt(3)-1

x = np.linspace(0,15,10000)

import matplotlib.pyplot as plt

plt.figure()
plt.plot(x,f(x), label=r"$f(x)$")
plt.plot(x,c(a_star)*g(a_star,x), label=r"$c(a^*)g_{a^*}(x)$")
plt.legend()
plt.show()
```
